```python
# 6 Treating outliers
box_columns= ['G1','G2','G3','absences']

q1 = df[box_columns].quantile(0.25)
q3 = df[box_columns].quantile(0.75)

iqr = q3-q1

lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

df_new = df[
    (df['G1'] >= lower['G1']) & (df['G1'] <= upper['G1']) &
    (df['G2'] >= lower['G2']) & (df['G2'] <= upper['G1']) &
    (df['G3'] >= lower['G3']) & (df['G3'] <= upper['G3']) &
    (df['absences'] >= lower['absences']) & (df['absences'] <= upper['absences'])
]
plt.figure(figsize=(10,6))

df_new[box_columns].plot(kind='box')
plt.title('Student Performance')
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
# 4 performing Binary mapping and One Hot Encoding
```

```python
#binary mapping

binary_map = {
    'yes':1, 'no':0,
    'M':1, 'F':0,
    'U':1, 'R':0,
    'T':1, 'A':0,
    'GT3':1, 'LE3':0,
    'GP':1, 'MS':0
}

binary_columns = ['school','sex','address','Pstatus','famsize',
               'schoolsup','famsup','paid','activities',
               'nursery','higher','internet','romantic']

for c in binary_columns:
    df[c] = df[c].map(binary_map)
```

```python
#One-hot encoding
df = pd.get_dummies(df, columns=['Mjob','Fjob','reason','guardian'], drop_first=True).astype
```

```python
# 8 Apply normalization and standardization

from sklearn.preprocessing import MinMaxScaler

numeric_cols = [
    'age', 'Medu', 'Fedu', 'traveltime', 'studytime',
    'failures', 'famrel', 'freetime', 'goout',
    'Dalc', 'Walc', 'health', 'absences',
    'G1', 'G2', 'G3',
    'average_grade', 'parents_education'
]

df_nor = df.copy()
scaler = MinMaxScaler()
df_nor[numeric_cols] = scaler.fit_transform(df_nor[numeric_cols])
```

```python

from sklearn.preprocessing import StandardScaler

df_stan = df.copy()
scaler = StandardScaler()
df_stan[numeric_cols] = scaler.fit_transform(df_stan[numeric_cols])
```

```python
# 9 split dataset into train/test
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
df['pass_fail'] = df['G3'].apply(lambda x: 1 if x >= 10 else 0)

X = df.drop(columns=['G3', 'pass_fail'])
y = df['pass_fail']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

```python
# 10. train a logistic regression model and exaluate

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
#prediction

predict = model.predict(X_test)
```

```python
predict
```

```python
#exaluate metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy", accuracy_score(y_test, predict))
print("\nClassification Report\n", classification_report(y_test, predict))
cm = confusion_matrix(y_test, predict)
print("\nConfusion Matrix\n", cm)
```
